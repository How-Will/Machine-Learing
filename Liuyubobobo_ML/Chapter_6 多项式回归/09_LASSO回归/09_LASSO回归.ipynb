### 比较Ridge和LASSO


Ridge，得到的模型依然是曲线。  
LASSO，得到的模型更接近直线。  
说明使用LASSO时，有很多特征前的系数为0。而使用Ridge时，有很多特征前面的系数不为0。  
即LASSO趋向于使得一部分theta值变为0。即这些特征是有用的。**所以LASSO可以作为特征选择用。**  
原理：  


LASSO有可能会错误将一些有用的特征变为0。故从计算精确度来说，还是岭回归更为准确。但是当特征特别大时，可以使用LASSO将特征变小。

### 数据准备

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.random.uniform(-3.0, 3.0, size=100)
X = x.reshape(-1, 1)
y = 0.5 * x + 3 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(x, y)
plt.show()
```

### 未添加正则化项

```python
from sklearn.model_selection import train_test_split

np.random.seed(666)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

def PolynomialRegression(degree):
    return Pipeline([
        ("poly", PolynomialFeatures(degree=degree)),
        ("std_scaler", StandardScaler()),
        ("lin_reg", LinearRegression())
    ])
```

```python
from sklearn.metrics import mean_squared_error

poly_reg = PolynomialRegression(degree=20)
poly_reg.fit(X_train, y_train)

y_predict = poly_reg.predict(X_test)
mean_squared_error(y_test, y_predict)  # 可以发现均方差特别大，即过拟合
```

```python
def plot_model(model):
    X_plot = np.linspace(-3, 3, 100).reshape(100, 1)
    y_plot = model.predict(X_plot)

    plt.scatter(x, y)
    plt.plot(X_plot[:,0], y_plot, color='r')
    plt.axis([-3, 3, 0, 6])
    plt.show()

plot_model(poly_reg)
```

### 使用LASSO回归

```python
from sklearn.linear_model import Lasso

def LassoRegression(degree, alpha):
    return Pipeline([
        ("poly", PolynomialFeatures(degree=degree)),
        ("std_scaler", StandardScaler()),
        ("lasso_reg", Lasso(alpha=alpha))
    ])
```

```python
lasso1_reg = LassoRegression(20, 0.01) # 这里的alpha取得值小于岭回归的alpha。是因为岭回归的正则化项是theta的平方，比较大
lasso1_reg.fit(X_train, y_train)

y1_predict = lasso1_reg.predict(X_test)
mean_squared_error(y_test, y1_predict)
```

```python
plot_model(lasso1_reg) # 模型平滑了
```

```python
lasso2_reg = LassoRegression(20, 0.1)
lasso2_reg.fit(X_train, y_train)

y2_predict = lasso2_reg.predict(X_test)
mean_squared_error(y_test, y2_predict)  # 均方差降低了
```

```python
plot_model(lasso2_reg)
```

```python
lasso3_reg = LassoRegression(20, 1)
lasso3_reg.fit(X_train, y_train)

y3_predict = lasso3_reg.predict(X_test)
mean_squared_error(y_test, y3_predict) # 正则化过头了
```

```python
plot_model(lasso3_reg)
```
