### 多项式回归


可以将x\*\*2看成是一个特征。
那么仍可以看成是线性方程

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size = 100)
X = x.reshape(-1, 1)  # 100个样本，每个样本1个特征
```

```python
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size = 100)
```

```python
plt.scatter(x, y)
plt.show()
```

### 使用线性回归

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, y_predict, color = 'r')
plt.show()
```

### 解决方案：添加一个特征

```python
(X**2).shape
```

```python
X2 = np.hstack([X, X**2])  # 每一个样本有两个特征,将x**2看作一个新的特征。那么从x的维度来看，则形成了一条曲线
X2.shape
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
y_predict2 = lin_reg2.predict(X2)
```

```python
plt.scatter(x, y)
#plt.plot(x, y_predict2, color = 'r') #不这么写的原因是x是乱序的，这么写会导致绘制出的图像不平滑
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color = 'r')
plt.show()
```

```python
lin_reg2.coef_  # (x, x**2)的系数
```

```python
lin_reg2.intercept_ # 截距
```
