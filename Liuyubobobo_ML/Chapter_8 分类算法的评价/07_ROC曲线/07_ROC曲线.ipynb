### ROC curve

描述TPR和FPR之间的关系  

TPR：本质就是预测为1，并且预测对的数量，占真实值为1的百分比是多少  
FPR：本质就是预测为1，但是预测错的数量，占真实值为0的百分比是多少  

以逻辑回归为例，通过改变不同的threshold来判断TPR和FPR之间的关系，可以看出TPR和FPR之间的关系是正相关的。  
下图中⭐代表预测值为1，●代表预测值为0


对于ROC曲线，通常关注的是曲线下方的面积，面积越大，表示算法性能越好


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

digits = datasets.load_digits()
X = digits.data
y = digits.target.copy()

y[digits.target==9] = 1
y[digits.target!=9] = 0
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
decision_scores = log_reg.decision_function(X_test)
```

```python
from playML.metric import FPR, TPR

fprs = []
tprs = []
thresholds = np.arange(np.min(decision_scores), np.max(decision_scores), 0.1)
for threshold in thresholds:
    y_predict = np.array(decision_scores >= threshold, dtype='int')
    fprs.append(FPR(y_test, y_predict))
    tprs.append(TPR(y_test, y_predict))
```

```python
plt.plot(fprs, tprs)
plt.show()
```

### scikit-learn中的ROC

```python
from sklearn.metrics import roc_curve

fprs, tprs, threshold = roc_curve(y_test, decision_scores)
```

```python
plt.plot(fprs, tprs)
plt.show()
```

### ROC AUC

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, decision_scores)
```
