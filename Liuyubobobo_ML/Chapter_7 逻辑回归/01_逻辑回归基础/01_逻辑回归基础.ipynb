### sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(t):
    return 1. / (1. + np.exp(-t))
```

```python
x = np.linspace(-10, 10, 500)

plt.plot(x, sigmoid(x))
plt.show()
```
