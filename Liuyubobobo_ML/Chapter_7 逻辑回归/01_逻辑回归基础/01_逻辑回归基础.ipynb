### 逻辑回归

目的：解决分类问题

** 回归问题怎么解决分类问题？ **  
** 将样本的特征和样本发生的概率联系起来，概率是一个数。所以可以叫做回归问题 **  

逻辑回归会计算出一个概率值p。之后根据概率值进行分类。上图中的1和0，对于不同的问题可能有不同的表示含义。  
对比KNN算法具有天生的解决多分类问题的性质，逻辑回归只能解决二分类问题。  


预测的结果范围[-inf, +inf]。于是需要对结果进行再次处理。解决方案：使用sigmoid

### sigmoid函数

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(t):
    return 1. / (1. + np.exp(-t))
```

```python
x = np.linspace(-10, 10, 500)

plt.plot(x, sigmoid(x))
plt.show()
```
